```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("tmdb/tmdb-movie-metadata")

print("Path to dataset files:", path)
```

```python
# Data description
import pandas as pd
import numpy as nu
import os
import json
# Load Data
md = pd.read_csv(os.path.join(path, 'tmdb_5000_movies.csv'))
md.head()
md.shape
```

```python
# Parse JSON Features
def extract_genres(genres_str):
  try:
    genres = json.loads(genres_str)
    return [genre['name'] for genre in genres]
  except:
    return []
md['genres'] = md['genres'].apply(extract_genres)
```

```python
def extract_keywords(keywords_str):
    try:
        keywords = json.loads(keywords_str)
        return [kw['name'] for kw in keywords]
    except:
        return []

md['keywords'] = md['keywords'].apply(extract_keywords)
```

```python
# Create combined_features for TF-IDF
def create_combined_features(row):
    genres = ' '.join(row['genres']) if isinstance(row['genres'], list) else ''
    keywords = ' '.join(row['keywords']) if isinstance(row['keywords'], list) else ''
    overview = row['overview'] if pd.notna(row['overview']) else ''

    return f"{genres} {keywords} {overview}"
```

```python
# Optional: Clean and normalize
md['combined_features'] = md.apply(create_combined_features, axis=1)
md['combined_features'] = md['combined_features'].str.lower()
```
